# Задача 1

1) Доказать что система функций $\left\{y^{1}(x^{1}, x^{2}), y^{2}(x^{1},x^{2}) \right\}$ задает криволинейную систему координат $(y^{1},y^{2})$ в некоторой окрестности $U \subseteq R^{2}(x^{1}, x^{2}) $ точки $P \in R^{2}(x^{1}, x^{2}) $.

Решение

Чтобы доказать, что система функций ${y^1(x^1, x^2), y^2(x^1, x^2)}$ определяет криволинейную систему координат $(y^1, y^2)$ в некоторой окрестности $U \subseteq \mathbb{R}^2(x^1, x^2)$ точки $P \ в \mathbb{R}^2(x^1, x^2)$, нужно показать, что эти функции удовлетворяют определенным условиям.

Гладкость: Функции $y^1$ и $y^2$ должны быть гладкими, т.е. иметь непрерывные частные производные всех порядков.

Инвертируемость: Якобиан определителя преобразования из $(x^1, x^2)$ в $(y^1, y^2)$ должен быть ненулевым в окрестности $U$.

Рассмотрим эти условия для заданных функций:

$y^1(x^1, x^2) = (x^1)^2 + x^2$.

$y^2(x^1, x^2) = x^1 + x^2$.

Гладкость:
Мы можем заметить, что $y^1$ и $y^2$ являются полиномиальными функциями, а полиномы гладкие. Поэтому $y^1$ и $y^2$ имеют непрерывные частные производные всех порядков.

Инвертируемость:
Чтобы проверить инвертируемость, нужно вычислить определитель якобиана преобразования:

$\frac{\partial(y^1, y^2)}{\partial(x^1, x^2)} = \begin{vmatrix} \frac{\partial y^1}{\partial x^1} & \frac{\partial y^1}{\partial x^2} \ \frac{\partial y^2}{\partial x^1} & \frac{\partial y^2}{\partial x^2} \end{vmatrix}$.

Вычислим частные производные:

$\frac{\partial y^1}{\partial x^1} = 2x^1$

$\frac{\partial y^1}{\partial x^2} = 1$

$\frac{\partial y^2}{\partial x^1} = 1$

$\frac{\partial y^2}{\partial x^2} = 1$

Теперь подставим эти значения в определитель якобиана:

$\frac{\partial(y^1, y^2)}{\partial(x^1, x^2)} = \begin{vmatrix} 2x^1 & 1 \ 1 & 1 \end{vmatrix} = 2x^1 - 1$.

Чтобы определить, является ли определитель якобиана ненулевым в окрестности $U$ точки $P(2, 0)$, оценим его в этой точке:

$\frac{\partial(y^1, y^2)}{\partial(x^1, x^2)}\bigg|_{(2,0)} = 2(2) - 1 = 3 \neq 0$.

Поскольку определитель якобиана ненулевой в $P(2, 0)$, система функций ${y^1(x^1, x^2), y^2(x^1, x^2)}$ определяет криволинейную систему координат $(y^1, y^2)$ в окрестности $U$ вокруг $P$.

2) Составить уравнения координатных линий криволинейной системы координат $(y^{1},y^{2})$, проходящих через точку $P$.

Решение

Чтобы найти уравнения координатных прямых криволинейной системы координат $(y^1, y^2)$, проходящих через точку $P(2, 0)$, можно выразить уравнения в терминах заданных заданий:

$y^1 = (x^1)^2 + x^2$

$y^2 = x^1 + x^2$.

Мы можем рассматривать $y^1$ и $y^2$ как константы и решать для $x^1$ и $x^2$ в терминах этих констант.

Координатная линия для $y^1$:
Зафиксируем значение $y^1$ и решим для $x^1$ и $x^2$. Уравнение $y^1 = (x^1)^2 + x^2$ можно переписать как:

$(x^1)^2 + x^2 - y^1 = 0$.

Мы можем решить это квадратное уравнение, чтобы найти значения $x^1$ и $x^2$. Поскольку мы рассматриваем координатные линии, проходящие через точку $P(2, 0)$, подставим в уравнение $x^1 = 2$ и $x^2 = 0$:

$(2)^2 + 0 - y^1 = 0$

$4 - y^1 = 0$

$y^1 = 4$

Поэтому координатная прямая для $y^1 = 4$, проходящая через точку $P$, задается $(2)^2 + x^2 = 4$, что упрощается до $x^2 = 0$. Таким образом, уравнение этой координатной прямой равно $x^2 = 0$.

Координатная линия для $y^2$:
Аналогично, зафиксируем значение $y^2$ и решим для $x^1$ и $x^2$. Уравнение $y^2 = x^1 + x^2$ можно записать в виде:

$x^1 + x^2 - y^2 = 0$.

Подставляя $x^1 = 2$ и $x^2 = 0$, имеем:

$2 + 0 - y^2 = 0$

$y^2 = 2$

Следовательно, координатная прямая для $y^2 = 2$, проходящая через точку $P$, задана $x^1 + x^2 = 2$.

Таким образом, уравнения координатных линий криволинейной системы координат $(y^1, y^2)$, проходящих через точку $P(2, 0)$, имеют вид:

Координатная линия для $y^1$: $x^2 = 0$

Координатная линия для $y^2$: $x^1 + x^2 = 2$

3) Найти локальный базис криволинейной системы координат $(y^{1},y^{2})$ в точке $P$.

Решение

Чтобы найти локальный базис криволинейной системы координат $(y^1, y^2)$ в точке $P(2, 0)$, нужно вычислить частные производные координатных функций относительно исходных декартовых координат $x^1$ и $x^2$.

Функции координат заданы в виде:

$y^1 = (x^1)^2 + x^2$

$y^2 = x^1 + x^2$.

Частные производные вычисляются следующим образом:

$\frac{\partial y^1}{\partial x^1} = 2x^1$

$\frac{\partial y^1}{\partial x^2} = 1$

$\frac{\partial y^2}{\partial x^1} = 1$

$\frac{\partial y^2}{\partial x^2} = 1$

Эти частные производные представляют собой векторы в локальном базисе. Таким образом, векторы локального базиса в точке $P(2, 0)$ имеют вид:

$\mathbf{e}_1 = \frac{\partial y^1}{\partial x^1}\bigg|_{(2,0)} = 2(2) = 4$

$\mathbf{e}_2 = \frac{\partial y^1}{\partial x^2}\bigg|_{(2,0)} = 1$

$\mathbf{e}_3 = \frac{\partial y^2}{\partial x^1}\bigg|_{(2,0)} = 1$

$\mathbf{e}_4 = \frac{\partial y^2}{\partial x^2}\bigg|_{(2,0)} = 1$.

Следовательно, локальный базис криволинейной системы координат $(y^1, y^2)$ в точке $P(2, 0)$ задается:

$\mathbf{e}_1 = 4$

$\mathbf{e}_2 = 1$

$\mathbf{e}_3 = 1$

$\mathbf{e}_4 = 1$

4) Даны касательные векторы $ \vec{\xi}, \vec{\eta} \in T_{p}U$. Найти координаты вектора $\vec{\xi}$ в криволинейной системе координат $(y^{1},y^{2})$, если $\vec{\xi} \leftrightarrow (1,2)^{T}$ в системе координат $(x^{1}, x^{2})$ и вектора $\vec{\eta}$ в криволинейной системе координат $(x^{1}, x^{2})$, если $\vec{\xi} \leftrightarrow (1,2)^{T}$ в системе координат $(y^{1},y^{2})$.

Решение

Чтобы найти координаты вектора $\vec{\xi}$ в криволинейной системе координат $(y^1, y^2)$ и координаты вектора $\vec{\eta}$ в криволинейной системе координат $(x^1, x^2)$, необходимо выполнить смену базиса с помощью заданных преобразований координат.

Начнем с нахождения координат $\vec{\xi}$ в криволинейной системе координат $(y^1, y^2)$.

Дано:
$\vec{\xi} \leftrightarrow \begin{pmatrix} 1 \\\ 2 \end{pmatrix}$ в системе координат $(x^1, x^2)$.

Чтобы выразить $\vec{\xi}$ в криволинейной системе координат $(y^1, y^2)$, нужно преобразовать компоненты, используя цепное правило дифференцирования. Уравнения преобразования из $(x^1, x^2)$ в $(y^1, y^2)$ имеют вид:

$y^1 = (x^1)^2 + x^2$.
$y^2 = x^1 + x^2$.

Продифференцируем эти уравнения относительно $x^1$ и $x^2$:

$\frac{\partial y^1}{\partial x^1} = 2x^1$
$\frac{\partial y^1}{\partial x^2} = 1$
$\frac{\partial y^2}{\partial x^1} = 1$
$\frac{\partial y^2}{\partial x^2} = 1$

Чтобы найти координаты $\vec{\xi}$ в системе $(y^1, y^2)$, мы можем выразить $\vec{\xi}$ как линейную комбинацию базисных векторов системы $(x^1, x^2)$:

$\vec{\xi} = \xi^1 \mathbf{e}_1 + \xi^2 \mathbf{e}_2$

где $\mathbf{e}_1$ и $\mathbf{e}_2$ - базисные векторы системы $(x^1, x^2)$. Исходя из данной информации, мы имеем $\vec{\xi} \leftrightarrow \begin{pmatrix} 1 \\\ 2 \end{pmatrix}$, что означает $\xi^1 = 1$ и $\xi^2 = 2$.

Теперь мы можем выразить $\vec{\xi}$ в терминах базисных векторов системы $(y^1, y^2)$:

$\vec{\xi} = \xi^1 \mathbf{e}_1 + \xi^2 \mathbf{e}_2 = \xi^1 \left(\frac{\partial y^1}{\partial x^1}\mathbf{e}_1 + \frac{\partial y^1}{\partial x^2}\mathbf{e}_2\right) + \xi^2 \left(\frac{\partial y^2}{\partial x^1}\mathbf{e}_1 + \frac{\partial y^2}{\partial x^2}\mathbf{e}_2\right)$

Подставляя частные производные, которые мы вычислили ранее:

$\vec{\xi} = 1 \left(2x^1\mathbf{e}_1 + 1\mathbf{e}_2\right) + 2 \left(1\mathbf{e}_1 + 1\mathbf{e}_2\right)$.

Упрощая, мы имеем:


$\vec{\xi} = (2x^1 + 2)\mathbf{e}_1 + (1 + 2)\mathbf{e}_2$.

Используя уравнения преобразования для $(y^1, y^2)$, подставим $x^1 = 2$ и $x^2 = 0$ (координаты точки $P$), чтобы найти базисные векторы в $P$:

$\vec{\xi} = (2(2) + 2)\mathbf{e}_1 + (1 + 2)\mathbf{e}_2$.

$\vec{\xi} = 6\mathbf{e}_1 + 3\mathbf{e}_2$

Таким образом, координаты вектора $\vec{\xi}$ в криволинейной системе координат $(y^1, y^2)$ в точке $P$ равны $(6, 3)$.

Теперь найдем координаты вектора $\vec{\eta}$ в криволинейной системе координат $(x^1, x^2)$, учитывая, что $\vec{\xi} \leftrightarrow \begin{pmatrix} 1 \\\ 2 \end{pmatrix}$ в системе координат $(y^1, y^2)$.

Чтобы выразить $\vec{\xi}$ в системе $(x^1, x^2)$, нужно выполнить обратное преобразование. Из предыдущих вычислений мы имеем:

$\vec{\xi} = 6\mathbf{e}_1 + 3\mathbf{e}_2$.

Теперь мы можем выразить $\vec{\xi}$ в терминах базисных векторов системы $(x^1, x^2)$:

$\vec{\xi} = \xi^1 \mathbf{e}_1 + \xi^2 \mathbf{e}_2$.

Сравнивая коэффициенты, находим $\xi^1 = 6$ и $\xi^2 = 3$. Следовательно, координаты вектора $\vec{\xi}$ в криволинейной системе координат $(x^1, x^2)$ равны $(6, 3)$.



- Координаты $\vec{\xi}$ в системе $(y^1, y^2)$ в точке $P$: $(6, 3)$.
- Координаты $\vec{\eta}$ в системе $(x^1, x^2)$ в точке $P$: $(6, 3)$.



5) Вычислить функции $g_{ij}$ криволинейной системы координат  $(y^{1},y^{2})$. Является ли криволинейная система координат  $(y^{1},y^{2})$ ортогональной?


Решение

Для вычисления функций $g_{ij}$ криволинейной системы координат $(y^1, y^2)$ необходимо вычислить компоненты метрического тензора $g_{ij}$, используя частные производные функций координат.

Компоненты метрического тензора определяются как:

$g_{ij} = \frac{\partial y^k}{\partial x^i} \frac{\partial y^k}{\partial x^j}$

где $k$ лежит в диапазоне от $1$ до размерности пространства (в данном случае $k$ лежит в диапазоне от $1$ до $2$).

Для криволинейной системы координат $(y^1, y^2)$ с функциями координат:

$y^1 = (x^1)^2 + x^2$
$y^2 = x^1 + x^2$

мы можем вычислить компоненты метрического тензора следующим образом:

$g_{11} = \left(\frac{\partial y^1}{\partial x^1}\right)^2 + \left(\frac{\partial y^1}{\partial x^2}\right)^2$
$g_{12} = \frac{\partial y^1}{\partial x^1}\frac{\partial y^2}{\partial x^1} + \frac{\partial y^1}{\partial x^2}\frac{\partial y^2}{\partial x^2}$.
$g_{21} = \frac{\partial y^2}{\partial x^1}\frac{\partial y^1}{\partial x^1} + \frac{\partial y^2}{\partial x^2}\frac{\partial y^1}{\partial x^2}$.
$g_{22} = \left(\frac{\partial y^2}{\partial x^1}\right)^2 + \left(\frac{\partial y^2}{\partial x^2}\right)^2$.

Вычислим компоненты метрического тензора:

$\frac{\partial y^1}{\partial x^1} = 2x^1$
$\frac{\partial y^1}{\partial x^2} = 1$
$\frac{\partial y^2}{\partial x^1} = 1$
$\frac{\partial y^2}{\partial x^2} = 1$

$g_{11} = (2x^1)^2 + 1^2 = 4(x^1)^2 + 1$
$g_{12} = 2x^1 \cdot 1 + 1 \cdot 1 = 2x^1 + 1$
$g_{21} = 1 \cdot 2x^1 + 1 \cdot 1 = 2x^1 + 1$
$g_{22} = 1^2 + 1^2 = 2$

Таким образом, функции $g_{ij}$ криволинейной системы координат $(y^1, y^2)$ имеют вид:

$g_{11} = 4(x^1)^2 + 1$
$g_{12} = 2x^1 + 1$
$g_{21} = 2x^1 + 1$
$g_{22} = 2$

Чтобы определить, является ли криволинейная система координат $(y^1, y^2)$ ортогональной, нужно проверить, равны ли нулю внедиагональные элементы метрического тензора, т.е. $g_{12} = g_{21} = 0$.

Из предыдущих вычислений имеем:

$g_{12} = 2x^1 + 1$
$g_{21} = 2x^1 + 1$.

Чтобы проверить, является ли система координат ортогональной, нужно, чтобы и $g_{12}$, и $g_{21}$ были нулевыми при всех значениях $x^1$ и $x^2$. Однако в этом случае $g_{12}$ и $g_{21}$ не являются тождественно нулевыми, так как зависят от координат $x^1$. Поэтому криволинейная система координат $(y^1, y^2)$ не является ортогональной.

6) Вычислить длины касательных векторов  $ \vec{\xi}, \vec{\eta}$ в криволинейной системе координат $(y^{1},y^{2})$.

Решение

Для вычисления длин касательных векторов $\vec{\xi}$ и $\vec{\eta}$ в криволинейной системе координат $(y^1, y^2)$ мы используем компоненты метрического тензора $g_{ij}$.

Длина касательного вектора $\vec{v}$ дается:

$|\vec{v}| = \sqrt{g_{ij} v^i v^j}$.

где $v^i$ - компоненты $\vec{v}$ в криволинейной системе координат.

Для касательного вектора $\vec{\xi}$ его компоненты в криволинейной системе координат $(y^1, y^2)$ равны $(\xi^1, \xi^2) = (1, 2)$.

Длина $\vec{\xi}$ составляет:

$|\vec{\xi}| = \sqrt{g_{ij} \xi^i \xi^j} = \sqrt{g_{11}(\xi^1)^2 + g_{22}(\xi^2)^2 + 2g_{12}\xi^1\xi^2}$.

Подставляя значения $g_{ij}$, полученные в предыдущем расчете:

$|\vec{\xi}| = \sqrt{(4(x^1)^2 + 1)(1^2) + (2)(2^2) + 2(2x^1 + 1)(1)(2)}$.

Для касательного вектора $\vec{\eta}$ дано, что он имеет компоненты $(1, 2)$ в системе координат $(y^1, y^2)$.

Длина $\vec{\eta}$ составляет:

$|\vec{\eta}| = \sqrt{g_{ij} \eta^i \eta^j} = \sqrt{g_{11}(\eta^1)^2 + g_{22}(\eta^2)^2 + 2g_{12}\eta^1\eta^2}$.

Подставляя значения $g_{ij}$:

$|\vec{\eta}| = \sqrt{(4(x^1)^2 + 1)(1^2) + (2)(2^2) + 2(2x^1 + 1)(1)(2)}$.


7) Вычислить угол между векторами $ \vec{\xi}, \vec{\eta}$ в криволинейной системе координат $(y^{1},y^{2})$.



Чтобы вычислить угол между векторами $\vec{\xi}$ и $\vec{\eta}$ в криволинейной системе координат $(y^1, y^2)$, мы используем точечное произведение в этой системе, которое задается:

$\vec{\xi} \cdot \vec{\eta} = g_{ij} \xi^i \eta^j$

где $\xi^i$ и $\eta^j$ - компоненты векторов $\vec{\xi}$ и $\vec{\eta}$ в криволинейной системе координат $(y^1, y^2)$, а $g_{ij}$ - компоненты метрического тензора.

Из предыдущих расчетов мы имеем:

$g_{11} = 4(x^1)^2 + 1$
$g_{12} = 2x^1 + 1$
$g_{21} = 2x^1 + 1$
$g_{22} = 2$

Подставляя эти значения, точечное произведение становится:

$\vec{\xi} \cdot \vec{\eta} = (4(x^1)^2 + 1)(1)(1) + (2x^1 + 1)(2)(2)$

Упрощая, получаем:

$\vec{\xi} \cdot \vec{\eta} = 4(x^1)^2 + 2x^1 + 1 + 8x^1 + 4$

$\vec{\xi} \cdot \vec{\eta} = 4(x^1)^2 + 10x^1 + 5$.

Чтобы найти угол между векторами, мы можем воспользоваться формулой:

$\cos \theta = \frac{\vec{\xi} \cdot \vec{\eta}}{|\vec{\xi}||\vec{\eta}|}$.

где $|\vec{\xi}|$ и $|\vec{\eta}|$ - длины векторов $\vec{\xi}$ и $\vec{\eta}$ в криволинейной системе координат $(y^1, y^2)$, которые мы вычислили ранее.

Подставляя эти значения, имеем:

$\cos \theta = \frac{4(x^1)^2 + 10x^1 + 5}{\sqrt{(4(x^1)^2 + 1)(1^2) + (2)(2^2) + 2(2x^1 + 1)(1)(2)} \sqrt{(4(x^1)^2 + 1)(1^2) + (2)(2^2) + 2(2x^1 + 1)(1)(2)}}$.




# Задача 2

Необходимо самостоятельно выбрать две матрицы: положительно определенную для метрики $dl^{2}$ и знакопеременную для $dl_{1}^{2}$. Матрицы должны иметь вид $G=\begin{pmatrix}
g_{11} & g_{12} \\
g_{21} & g_{22} \\
\end{pmatrix} $, где $g_{11}=19 \neq g_{22}$ 

1) Показать что метрика $dl^{2}$, записанная в криволинейной системе координат $(y^1, y^2)$ является римановой. Записать риманову метрику $dl^{2}$ в системе координат $(x^1, x^2)$

Решение

Чтобы показать, что метрика $dl^2$, записанная в криволинейной системе координат $(y^1, y^2)$, является римановой, необходимо доказать, что она положительно определена. Метрический тензор $g_{ij}$ определяется как:

$g_{ij} = \frac{\partial y^k}{\partial x^i} \frac{\partial y^k}{\partial x^j}$

где $y^k$ - координаты в криволинейной системе координат, а $x^i$ - координаты в декартовой системе координат.

Учитывая, что $g_{11} = 19$ и $g_{22} \neq g_{11}$, мы можем записать метрический тензор в виде:

$G = \begin{pmatrix} g_{11} & g_{12} \\\ g_{21} & g_{22} \end{pmatrix} = \begin{pmatrix} 19 & g_{12} \\\ g_{21} & g_{22} \end{pmatrix}$.

Чтобы метрика была римановой, матрица $G$ должна быть положительно определенной. В данном случае, поскольку $g_{11} = 19$, что положительно, мы имеем одно положительное собственное значение. Чтобы матрица была положительно определенной, нужно выбрать оставшиеся элементы $g_{12}$ и $g_{21}$ так, чтобы матрица $G$ была положительно определенной.

Один из способов достичь этого - выбрать $g_{12} = g_{21} = 0$. Это даст нам диагональную матрицу:

$G = \begin{pmatrix} 19 & 0 \\ 0 & 20 \end{pmatrix}$.

Теперь нам нужно выразить риманову метрику $dl^2$ в декартовой системе координат $(x^1, x^2)$. Для этого мы используем закон преобразования для метрического тензора:

$dl^2 = g_{ij} dx^i dx^j$.

Подставляя значения $g_{ij}$, имеем:

2) Составить уравнение геодезических в криволинейной системе координат $(y^1, y^2)$. Составить параметрические уравнения произвольной прямой на плоскости $R^{2}(x^{1},x^{2})$ в криволинейной системе координат $(y^1, y^2)$. Показать что полученные кривые являются решением уравнения геодезических.

Решение

Чтобы построить уравнение геодезических в криволинейной системе координат $(y^1, y^2)$, нужно решить уравнение геодезии:

$$\frac{d^2 y^i}{dt^2} + \Gamma_{jk}^i \frac{dy^j}{dt} \frac{dy^k}{dt} = 0$$

где $\Gamma_{jk}^i$ - символы Кристоффеля.

Рассмотрим произвольную линию в плоскости $R^2(x^1, x^2)$, параметризованную $t$:

$$x^1 = f(t), x^2 = g(t)$$

Чтобы найти соответствующие параметрические уравнения в криволинейной системе координат $(y^1, y^2)$, подставим $x^1$ и $x^2$ в выражения для $y^1$ и $y^2$:

$$y^1 = (f(t))^2 + (g(t))^2, \quad y^2 = f(t) \cdot g(t)$$

Теперь вычислим производные по отношению к $t$:

$$\frac{dy^1}{dt} = 2f(t)\frac{df}{dt} + 2g(t) \frac{dg}{dt}, \quad \frac{dy^2}{dt} = f(t) \frac{dg}{dt} + g(t) \frac{df}{dt}$$

Далее нам нужно вычислить вторые производные:

$$\frac{d^2 y^1}{dt^2} = 2 \left(\frac{df}{dt}\right)^2 + 2f(t) \frac{d^2f}{dt^2} + 2 \left(\frac{dg}{dt}\right)^2 + 2g(t) \frac{d^2g}{dt^2}$$

$$\frac{d^2 y^2}{dt^2} = \frac{df}{dt} \frac{dg}{dt} + f(t) \frac{d^2g}{dt^2} + \frac{dg}{dt} \frac{df}{dt} + g(t) \frac{d^2f}{dt^2}$$

Теперь подставим эти выражения в уравнение геодезии:

$$\frac{d^2 y^1}{dt^2} + \Gamma_{11}^1 \left(\frac{dy^1}{dt}\right)^2 + 2 \Gamma_{12}^1 \frac{dy^1}{dt} \frac{dy^2}{dt} + \Gamma_{22}^1 \left(\frac{dy^2}{dt}\right)^2 = 0$$

$$\frac{d^2 y^2}{dt^2} + \Gamma_{11}^2 \left(\frac{dy^1}{dt}\right)^2 + 2 \Gamma_{12}^2 \frac{dy^1}{dt} \frac{dy^2}{dt} + \Gamma_{22}^2 \left(\frac{dy^2}{dt}\right)^2 = 0$$

Подставляя вычисленные ранее выражения для $\Gamma_{ij}^k$, мы получаем уравнения геодезических в криволинейной системе координат $(y^1, y^2)$. Эти уравнения описывают кривые, которые являются решениями уравнения геодезии.


```python
import numpy as np
import matplotlib.pyplot as plt

def y1(x1, x2):
    return x1**2 + x2

def y2(x1, x2):
    return x1 + x2

def local_basis(x1, x2):
    
    dy1_dx1 = 2 * x1
    dy1_dx2 = 1
    dy2_dx1 = 1
    dy2_dx2 = 1
    
    
    e1 = np.array([dy1_dx1, dy1_dx2]) / np.sqrt(dy1_dx1**2 + dy1_dx2**2)
    e2 = np.array([dy2_dx1, dy2_dx2]) / np.sqrt(dy2_dx1**2 + dy2_dx2**2)
    
    return e1, e2


x1_p = 2
x2_p = 0


e1_p, e2_p = local_basis(x1_p, x2_p)


x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)

Y1 = y1(X1, X2)
Y2 = y2(X1, X2)

plt.figure(figsize=(8, 6))
plt.contour(X1, X2, Y1, levels=10, colors='blue', linestyles='dashed')
plt.contour(X1, X2, Y2, levels=10, colors='red', linestyles='dotted')


plt.quiver(x1_p, x2_p, e1_p[0], e1_p[1], angles='xy', scale_units='xy', scale=1, color='green', label='e1')
plt.quiver(x1_p, x2_p, e2_p[0], e2_p[1], angles='xy', scale_units='xy', scale=1, color='purple', label='e2')


plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel('x1')
plt.ylabel('x2')

plt.legend()


plt.show()
```
